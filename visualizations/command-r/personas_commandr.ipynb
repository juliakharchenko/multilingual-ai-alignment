# SCATTERPLOT VISUALIZATIONS

## Individualism vs Collectivism

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the input file
file_path = "/content/output_commandr_personas_individualistic_collectivist.csv"
data = pd.read_csv(file_path)


# Calculate the percentage of individualistic responses for each country
individualistic_percentage = (data[data['Matching_Choice'] == 'Individualistic'].groupby('Target_Nationality').size() / data.groupby('Target_Nationality').size()) * 100

import scipy
print(scipy.stats.pearsonr(data['Individualistic_Collectivist_Score'], individualistic_percentage[data['Target_Nationality']], alternative='two-sided', method=None))

# Plot the scatterplot with linear regression
plt.figure(figsize=(10, 6))
sns.regplot(x=data['Individualistic_Collectivist_Score'], y=individualistic_percentage[data['Target_Nationality']], ci=None)
plt.title('Country Individualistic Value vs. \n% Individualistic Responses', fontsize=25)
plt.xlabel('Country Individualistic Value', fontsize=25)
plt.ylabel('% Individualistic Responses', fontsize=25)
plt.grid(True)
plt.show()
```

#WIN RATES

## INDIVIDUALISM VS COLLECTIVISM

```python
file_path = "/content/output_commandr_personas_individualistic_collectivist.csv"
data = pd.read_csv(file_path)

# Calculate the win rate (percentage of individualistic responses) for each country
win_rate = (data[data['Matching_Choice'] == 'Individualistic'].groupby('Target_Nationality').size() / data.groupby('Target_Nationality').size()) * 100

# Sort the win rate values for better visualization
win_rate = win_rate.sort_values()
print(len(win_rate))

collectivist_countries = ['Russia', 'China', 'Indonesia', 'Turkey', 'Poland', 'Iran',
                          'Romania', 'Thailand', 'Bulgaria', 'Kazakhstan', 'Armenia',
                          'Georgia', 'Albania', 'Azerbaijan', 'Malaysia', 'Mongolia',
                          'Belarus', 'India', 'South Africa', 'Sri Lanka']  # List of collectivist countries

# Define colors based on whether the target nationalities are individualistic or collectivist
colors = []
for country in win_rate.index:
    if country in collectivist_countries:
        colors.append('green')
    else:
        colors.append('orange')

# Plot the horizontal bar plot
plt.figure(figsize=(10, 8))
win_rate.plot(kind='barh', color=colors)
plt.title('Individualistic Response Rate by Country')
plt.xlabel('Individualistic Response Rate (%)')
plt.ylabel('Country')
legend_labels = ['Collectivist Country', 'Individualistic Country']
legend_colors = ['green', 'orange']
legend_handles = [plt.Rectangle((0,0),1,1, color=color) for color in legend_colors]
plt.legend(legend_handles, legend_labels, loc='upper left', bbox_to_anchor=(1, 1.1))
plt.grid(axis='x')
# Add win rate percentages on each bar
for i, v in enumerate(win_rate):
    plt.text(v + 0.5, i, f'{v:.2f}%', color='black', va='center')
plt.show()
```

# % RESPONSE ANALYSIS

## INDIVIDUALISM VS COLLECTIVISM

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the input file
file_path = "/content/output_commandr_personas_individualistic_collectivist.csv"
data = pd.read_csv(file_path)

# Calculate the counts of low and high power distance responses for each country
grouped_data = data.groupby(['Target_Nationality', 'Matching_Choice']).size().unstack(fill_value=0)

# Calculate the total number of responses for each country
grouped_data['Total'] = grouped_data.sum(axis=1)

# Calculate the proportion of low and high power distance responses for each country
grouped_data['Individualistic (%)'] = (grouped_data['Individualistic'] / grouped_data['Total']) * 100
grouped_data['Collectivist (%)'] = (grouped_data['Collectivist'] / grouped_data['Total']) * 100

# Sort the data by the proportion of low power distance responses for better visualization
grouped_data = grouped_data.sort_values(by='Individualistic (%)')

# Plot the stacked bar plot
plt.figure(figsize=(15, 8))  # Increase figure width for more space
grouped_data[['Individualistic (%)', 'Collectivist (%)']].plot(kind='barh', stacked=True, color=['green', 'orange'], width=0.4)  # Reduce bar width
plt.xlabel('Percentage (%)')
plt.ylabel('Country')
plt.title('Comparison of Individualist vs Collectivist Responses by Country')
plt.legend(title='Matching Choice', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='x')
plt.tick_params(axis='y', which='major', labelsize=8)
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

# HISTOGRAM DISTRIBUTIONS

## INDIVIDUALISM VS COLLECTIVISM

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data from the input file
file_path = "/content/output_commandr_personas_individualistic_collectivist.csv"
data = pd.read_csv(file_path)

# Separate the data into Masculine and Feminine responses
masculine_data = data[data['Matching_Choice'] == 'Individualistic']
feminine_data = data[data['Matching_Choice'] == 'Collectivist']

# Get unique country names and their counts
unique_countries = data['Target_Nationality'].unique()
num_countries = len(unique_countries)

# Create evenly spaced bins for each country
bins = np.arange(num_countries + 0.5) - 0.5

# Plot histograms for Masculine and Feminine responses
plt.figure(figsize=(12, 6))

# Histogram for Masculine responses
plt.hist(masculine_data['Target_Nationality'], bins=bins, alpha=0.7, color='green', label='Individualistic')

# Histogram for Feminine responses
plt.hist(feminine_data['Target_Nationality'], bins=bins, alpha=0.7, color='orange', label='Collectivist')

plt.xlabel('Country')
plt.ylabel('Frequency')
plt.title('Distribution of Individualistic and Collectivist Responses by Country')
plt.xticks(np.arange(num_countries), unique_countries, rotation=45)  # Set custom x-axis ticks
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# DENSITY CURVES TO SHOW DIFFERENTIATION OF VALUES

## INDIVIDUALISM VS COLLECTIVISM

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the data from the input file
file_path = "/content/output_commandr_personas_individualistic_collectivist.csv"
# file_path = "julia/output_openai_personas_individualistic_collectivist (2).csv"
data = pd.read_csv(file_path)
# collectivist_country_data['Matching_Choice'].value_counts()
# Individualistic    2562
# Collectivist       2357
# Inconclusive         81
# throw out inconclusive
data = data[data['Matching_Choice'] != 'Inconclusive']
# make Individualistic_Response a boolean
data['Individualistic_Response'] = data['Matching_Choice'] == 'Individualistic'

# aggregate by Target_Nationality and take the mean of Individualistic_Response
data = data.groupby('Target_Nationality').agg({'Individualistic_Response': 'mean'}).reset_index()

# Define lists of collectivist and individualistic countries
collectivist_countries = ['Russia', 'China', 'Indonesia', 'Turkey', 'Poland', 'Iran',
                          'Romania', 'Thailand', 'Bulgaria', 'Kazakhstan', 'Armenia',
                          'Georgia', 'Albania', 'Azerbaijan', 'Malaysia', 'Mongolia',
                          'Belarus', 'India', 'South Africa', 'Sri Lanka']

data['collectivist'] = data['Target_Nationality'].isin(collectivist_countries)


# Calculate frequencies of responses for each nationality
collectivist_country_perc = data[data['collectivist']]['Individualistic_Response']
individualistic_country_counts = data[~data['collectivist']]['Individualistic_Response']

# make KDE
plt.figure(figsize=(12, 6))
sns.kdeplot(collectivist_country_perc, color='green', label='Collectivist Countries', bw_adjust=0.5)
sns.kdeplot(individualistic_country_counts, color='orange', label='Individualistic Countries', bw_adjust=0.5)
plt.xlabel('Average % of Individualistic Responses (Country)', fontsize=25)
plt.ylabel('Density', fontsize=25)
plt.title('Command R Plus, Personas - Density of Responses by Frequency\n for Collectivist and Individualistic Countries', fontsize=25)
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import accuracy_score
import numpy as np

X = np.concatenate((collectivist_country_perc, individualistic_country_counts)).reshape(-1, 1)
y = np.concatenate((np.zeros(len(collectivist_country_perc)), np.ones(len(individualistic_country_counts))))
# sweep a dense linspace of thresholds
threshs = np.linspace(0, 1, 100)
# calculate the accuracy for each threshold
best_acc = max(
    np.max([accuracy_score(y, X.ravel() > thresh) for thresh in threshs]),
    np.max([accuracy_score(y, X.ravel() < thresh) for thresh in threshs])
)
# plt.plot(threshs, accuracies)
print(best_acc)
```

# COMPARISON BY COUNTRY'S RESOURCE LEVEL

## INDIVIDUALISM VS COLLECTIVISM

```python
# Separate the data into three groups based on the 'Resource' column
file_path = "/content/output_commandr_personas_individualistic_collectivist.csv"
data = pd.read_csv(file_path)
high_resource_data = data[data['Resource_Level'] == 'High']
mid_resource_data = data[data['Resource_Level'] == 'Mid']
low_resource_data = data[data['Resource_Level'] == 'Low']

# Group the data by 'Target_Nationality' and 'Adheres to Value', and count the occurrences for High resource
grouped_data_high = high_resource_data.groupby(['Target_Nationality', 'Adheres_to_Value']).size().unstack().fillna(0)

# Group the data by 'Target_Nationality' and 'Adheres to Value', and count the occurrences for Mid resource
grouped_data_mid = mid_resource_data.groupby(['Target_Nationality', 'Adheres_to_Value']).size().unstack().fillna(0)

# Group the data by 'Target_Nationality' and 'Adheres to Value', and count the occurrences for Low resource
grouped_data_low = low_resource_data.groupby(['Target_Nationality', 'Adheres_to_Value']).size().unstack().fillna(0)

# Plot the data for High resource
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

grouped_data_high.plot(kind='barh', stacked=True, ax=axes[0])
axes[0].set_xlabel('Count')
axes[0].set_ylabel('Country')
axes[0].set_title('Resource = High')
axes[0].legend(title='Adheres to Value')

# Plot the data for Mid resource
grouped_data_mid.plot(kind='barh', stacked=True, ax=axes[1])
axes[1].set_xlabel('Count')
axes[1].set_ylabel('Country')
axes[1].set_title('Resource = Mid')
axes[1].legend(title='Adheres to Value')

# Plot the data for Low resource
grouped_data_low.plot(kind='barh', stacked=True, ax=axes[2])
axes[2].set_xlabel('Count')
axes[2].set_ylabel('Country')
axes[2].set_title('Resource = Low')
axes[2].legend(title='Adheres to Value')

plt.tight_layout()
plt.show()
```
